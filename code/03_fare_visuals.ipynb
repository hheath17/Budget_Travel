```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('../data/flight_data_clean.csv')
```

```python
df.head()
```

```python
price_per_city1 = pd.DataFrame(df.groupby('city1')['fare'].sum().reset_index())

price_per_city1 = price_per_city1[price_per_city1['fare'] > 50]
```

```python
price_per_city1 = price_per_city1.sort_values(by = 'fare', ascending = False)
```

```python
#histogram of fares
plt.figure(figsize=(10, 6))
plt.hist(df['fare'], bins=30, edgecolor='k', alpha=0.7, color='purple')
plt.title('Distribution of Flight Fares', size=25)
plt.xlabel('Fare', size=15)
plt.ylabel('Frequency', size=15)
plt.grid(False)


#save and export
plt.tight_layout()
plt.savefig('../images/histo_flight_fares.png')

plt.show();
```

```python
# Calculate average fare by destination city
average_fares = df.groupby('city2')['fare'].mean().sort_values()

# Bar chart
plt.figure(figsize=(12, 8))
average_fares.plot(kind='barh', edgecolor='k', color='skyblue')
plt.title('Average Flight Fares by Destination City', size=20)
plt.xlabel('Average Fare', size= 15)
plt.ylabel('Destination City', size= 15)
plt.grid(False)

plt.tight_layout()
plt.savefig('../images/avg_fare_by_city.png')

plt.show();
```

```python
#box plot of fares by destination city
plt.figure(figsize=(14, 8))
df.boxplot(column='fare', by='city2', vert=False, patch_artist=True, color='red')
plt.title('Box Plot of Flight Fares by Destination City', size=18)
plt.xlabel('Fare', size=15)
plt.ylabel('Destination City', size=15)
plt.suptitle('')  
plt.grid(False)

plt.tight_layout()
plt.savefig('../images/box_plot_fare_by_city.png')

plt.show();
```
